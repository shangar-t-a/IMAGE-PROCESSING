# BASICS OF IMAGE PROCESSING USING PYTHON PART 1 (matplotlib)

__IMAGE MANIPULATION with Matplotlib__

    1) import necessary libraries

    2) inspect necessary information

    3) show the image

```python
from matplotlib import image, pyplot

FilePath = 'rose.jpg'
img = image.imread(FilePath)

print('Image dtype: ', img.dtype)
print('Image shape: ', img.shape)

pyplot.imshow(img)
```

__ROTATE , CROP , BRIGHTNESS AND CONTRAST__

    1) Image crop with PIL
    2) Looking at the pixel locations in above image, we can define the coordinate (box) to crop

```python
from PIL import Image

im = Image.open(FilePath)

Rotation_Degree = 30
#im.rotate(Rotation_Degree).show()
rot = im.rotate(Rotation_Degree)
pyplot.imshow(rot)
```

```python
box = (100, 50, 175, 125) #left, upper, right, lower
#im.crop(box).show()
crop = im.crop(box)
pyplot.imshow(crop)
```

# Image Enhancements with PIL: CONTRAST

```python
from PIL import ImageEnhance
enh = ImageEnhance.Contrast(im)
enh.enhance(1.75).show("75% more contrast")
```

# Image Enhancements with PIL: BRIGHTNESS

```python
enh = ImageEnhance.Brightness(im)
enh.enhance(1.5).show("50% more brightness")
```
